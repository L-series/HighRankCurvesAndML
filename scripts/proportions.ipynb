```python
import pandas as pd
from sympy import primepi

num = 0
primes = 0
ratio = []
for i in range(0,400):
    s = str(i).zfill(5)
    df = pd.read_csv(f'/run/media/eva/BACKUP/primefiles/pf/prime_all_{s}', names = ['cond','a1','a2','a3','a4','a6','sign'])
    df['cond'] = df['cond'].str.replace("[","")
    
    num += len(df['cond'].unique())
    primes += primepi((i+1)*10**9) 
    ratio.append(num/primes)

    
print(ratio)
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import seaborn as sns

# Define the curve function
def curve_function(x, a, b):
    return x**a * np.log(x)**b

# Create some example data (x_data, y_data)
x_data = [(x+1)*10**9 for x in range(0,362) ]
y_data = ratio

# Perform curve fitting
params, params_covariance = curve_fit(curve_function, x_data, y_data, p0=[1, 1])

# Print the best-fit parameters
print("a =", params[0], ", b =", params[1])
sns.set(style="whitegrid")
# Plot the data and the best-fit curve
x_fit = [(x+1)*10**9 for x in range(0,362)]
y_fit = curve_function(x_fit, params[0], params[1])

plt.scatter(x_data, y_data, label='Data', color='red')
plt.plot(x_fit, y_fit, label='Best-fit curve', color='blue')
plt.legend()
plt.show()
```
