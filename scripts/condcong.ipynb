```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sympy import primerange

# Initialize dictionaries for each prime
primes = list(primerange(3, 100))
# primes = [4,6,8,12]
prime_dicts = {prime: {} for prime in primes}

r = [x for x in range(1,400)] 

# Iterate through the files once and update the dictionaries
for i in r:
    file = '/run/media/eva/BACKUP/primefiles/pf/prime_all_' + str(i).zfill(5)
    df = pd.read_csv(file, names = ['cond','a1','a2','a3','a4','a6','sign'])
    df['cond'] = df['cond'].str.replace('[','').astype(int)

    
    for prime in primes:
        df[f'cond_mod{prime}'] = df['cond'] % prime
        conds = df[f'cond_mod{prime}'].value_counts().to_dict()
        
        # Update the value_counts dictionary with the new counts
        for key, value in conds.items():
            if key in prime_dicts[prime]:
                prime_dicts[prime][key] += value
            else:
                prime_dicts[prime][key] = value

```

```python
import itertools
import math 
import numpy as np

def generate_bar_chart(prime_dict, modulus,ax):
    condval = prime_dict

    # Prepare the color palette
    if modulus % 12 in [5, 11]:
        # If modulus is 5 or 11 mod 12, use a neutral color for all bars
        color_map = {i: 'gray' for i in range(1, modulus)}
    else:
        # If modulus is 7 or 1 mod 12, color the graph based on cosets
        num_colors = 6
        color_palette = sns.color_palette("deep", num_colors)        
        # Calculate the sixth powers modulo p
        sixth_powers = {pow(x, 6, modulus) for x in range(1, modulus)}

        # Calculate the cosets of the group of sixth powers
        cosets = []
        for i in range(1, modulus):
            coset = sorted({(i * elem) % modulus for elem in sixth_powers})
            if coset not in cosets:
                cosets.append(coset)

        # Create a color_map dictionary based on the cosets
        color_map = {}
        for i, coset in enumerate(cosets):
            for elem in coset:
                color_map[elem] = color_palette[i % num_colors]

        # If modulus is 7 or greater, color the most frequent bar and its pair column in bright red
        if (modulus >= 19):
            most_freq_key = max(condval, key=condval.get)
            color_map[most_freq_key] = 'red'
            color_map[modulus - most_freq_key] = 'red'

    colors = [color_map[key] for key in condval.keys()]

    # Plot a bar chart of the frequencies using seaborn
    sns.set(style="whitegrid")
    fig_width = 10
 #   fig, ax = plt.subplots(figsize=(fig_width, 6))

    # Convert the keys to strings for better display
    x_labels = [str(k) for k in condval.keys()]

    condval = dict(sorted(condval.items(), key=lambda x: x[1], reverse=True))

    # Plot the bar chart with custom colors
    sns.barplot(x=x_labels, y=list(condval.values()), ax=ax, palette=colors)

    # Set axis labels
#     ax.set_xlabel("Conductor mod" + str(modulus))
    ax.set_ylabel("Frequency")

    # Rotate the x-axis labels for better display
    plt.xticks(rotation=45)

    # Show the plot
 #   plt.show()

def plot_bar_charts_in_grid(primes, prime_dicts, congruence_class):
    filtered_primes = [prime for prime in primes if prime % 12 == congruence_class]
    num_primes = len(filtered_primes)
    num_cols = 2
    num_rows = math.ceil(num_primes / num_cols)

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 6), constrained_layout=True)

    for i, prime in enumerate(filtered_primes):
        row, col = divmod(i, num_cols)
        generate_bar_chart(prime_dicts[prime], prime, axes[ col])
        axes[ col].set_title(f"Redidue classes of prime conductors mod {prime}")

    # Remove extra subplots if any
    for j in range(i + 1, num_rows * num_cols):
        row, col = divmod(j, num_cols)
        fig.delaxes(axes[row, col])

    plt.show()


# Plot bar charts in grids for each congruence class modulo 12
for congruence_class in [1, 5, 7, 11]:
    print(f"Bar charts for prime moduli that are {congruence_class} mod 12")

    plot_bar_charts_in_grid(primes, prime_dicts, congruence_class)
```

```python
import itertools

def generate_bar_chart(prime_dict, modulus):
    condval = prime_dict

    # Calculate the total number of elements
    total_elements = sum(condval.values())

    # Calculate the relative frequencies
    relative_freq = {k: v / total_elements for k, v in condval.items()}
    cosets = []
    # Prepare the color palette
    if modulus % 12 in [5, 11]:
        # If modulus is 5 or 11 mod 12, use a neutral color for all bars
        color_map = {i: 'gray' for i in range(1, modulus)}
    else:
        # If modulus is 7 or 1 mod 12, color the graph based on cosets
        num_colors = 6
        color_palette = sns.color_palette("husl", num_colors)
        
        # Calculate the sixth powers modulo p
        sixth_powers = {pow(x, 6, modulus) for x in range(1, modulus)}

        # Calculate the cosets of the group of sixth powers
        
        for i in range(1, modulus):
            coset = sorted({(i * elem) % modulus for elem in sixth_powers})
            if coset not in cosets:
                cosets.append(coset)

        # Create a color_map dictionary based on the cosets
        color_map = {}
        for i, coset in enumerate(cosets):
            for elem in coset:
                color_map[elem] = color_palette[i % num_colors]

    colors = [color_map[key] for key in relative_freq.keys()]

    # Plot a bar chart of the frequencies using seaborn
    sns.set(style="whitegrid")
    fig_width = 6 + 0.15 * modulus
    fig, ax = plt.subplots(figsize=(fig_width, 6))

    # Convert the keys to strings for better display
    x_labels = [str(k) for k in relative_freq.keys()]

    relative_freq = dict(sorted(relative_freq.items(), key=lambda x: x[1], reverse=True))

    # Plot the bar chart with custom colors
    sns.barplot(x=x_labels, y=list(relative_freq.values()), ax=ax, palette=colors)

    # Set axis labels
    ax.set_xlabel("Conductor mod" + str(modulus))
    ax.set_ylabel("Relative Frequency")

    # Rotate the x-axis labels for better display
    plt.xticks(rotation=45)

    # Show the plot
    plt.show()
    coset_freqs = []
    for coset in cosets:
        coset_frequency = sum(relative_freq.get(elem, 0) for elem in coset)
        coset_freqs.append(coset_frequency)

    # Return the cosets and their frequencies for further processing
    print(cosets)
    print(coset_freqs)
    return cosets, coset_freqs

cosets_list = []
# Generate bar charts for each prime
for prime in primes:
    cosets = generate_bar_chart(prime_dicts[prime], prime)
    cosets_list.append(cosets)
```
