```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Initialize variables for storing partial results
total_rk = 0
count = 0
all_conds = []
all_avgs = []

# Process each file one at a time
for i in range(10):
    filename = f'~/results/trueranks/ranks_'+str(i).zfill(5)+'.csv'
    df = pd.read_csv(filename, names=['rk', 'cond', 'a1', 'a2', 'a3', 'a4', 'a6', 'sign'])

    # Calculate the running average for the current file
    for index, row in df.iterrows():
        total_rk += row['rk']
        count += 1
        running_avg = total_rk / count
        all_conds.append(row['cond'])
        all_avgs.append(running_avg)

```

```python


# Plot the accumulated running average
plt.figure(figsize=(10, 6))
plt.plot(all_conds[10000:], all_avgs[10000:])
plt.xlabel('cond')
plt.ylabel('Running Average of rank')
plt.title('Running Average of rank across all files')
plt.show()
```
