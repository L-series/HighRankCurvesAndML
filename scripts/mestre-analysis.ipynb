```python
y_val = []
import itertools
import numpy as np

for rank in range(0,20):

    # Read the list from the text file
    with open(f'/home/eva/results/scripts/sizes{rank}.txt', "r") as file:
        data = file.read().split()

    # Convert the list elements to integers
    data = [int(x) for x in data]
    print(rank)
    # Iterate over the list and replace zeros
    i = 0
    while i < len(data):
        if data[i] == 0:
            left = i - 1
            right = i + 1

            # Find the first non-zero element to the left
            while left >= 0 and data[left] == 0:
                left -= 1

            # Find the first non-zero element to the right
            while right < len(data) and data[right] == 0:
                right += 1

            # Calculate the average and replace zeros
            if left >= 0 and right < len(data):
                avg = (data[left] + data[right]) / 2
                for j in range(left + 1, right):
                    data[j] = avg
                i = right
            else:
                i += 1
        else:
            i += 1

    y_val = y_val + data
    
ydata = list(itertools.accumulate(y_val))
xs = [x for x in range(0,20000)]
np.savetxt('ydata.csv', ydata, delimiter=',')
```
